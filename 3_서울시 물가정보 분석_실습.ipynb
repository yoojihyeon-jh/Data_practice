# 삼겹살 가격 분석

## 삼겹살 데이터
* 특정 문자열이 포함되어있는지 확인 : .str.contains(문자열)

```python
df.head(2)
```

```python
# 2021-06 데이터 이용
df_sam = df[(df['품목 이름'].str.contains('삼겹살')) & (df['년도-월'] == '2021-06') & (df['실판매규격'].str.contains('600g'))]
df_sam
```

```python
# 삼겹살 600g의 평균 가격은?
df_sam['가격(원)'].mean()
```

```python
# 삼겹살 600g의 최고 가격은?
df_sam['가격(원)'].max()
```

```python
df_sam[df_sam['가격(원)']== df_sam['가격(원)'].max()]
```

```python
# 삼겹살 600g의 최저 가격은?
df_sam['가격(원)'].min()
```

```python
df_sam[df_sam['가격(원)']== df_sam['가격(원)'].min()]
```

## 우리동네 삼겹살 가격

```python
# 우리구 삼겹살 가격  시장/마트 이름, 품목 이름, 실판매규격' 가격(원)
gu = input('구 이름:  ')
df_market1 = df_sam[df_sam['자치구 이름'] == gu][['시장/마트 이름', '품목 이름', '실판매규격', '가격(원)']].drop_duplicates()
df_market1
```

```python
# 시각화
x = df_market1['시장/마트 이름']
y = df_market1['가격(원)']
plt.scatter(x,y)
plt.show()
```

## 마트 지점별 삼겹살 가격

```python
mart = input('시장/마트 이름: ') # 입력은 백화점만
```

```python
# 마트 지점별 삼겹살 가격
df_sam_mart = df_sam[df_sam['시장/마트 이름'].str.contains(mart)][['시장/마트 이름', '품목 이름', '실판매규격', '가격(원)']].drop_duplicates()
df_sam_mart
```

```python
# 시각화
x = df_sam_mart['시장/마트 이름']
y = df_sam_mart['가격(원)']
plt.bar(x,y)
plt.show()
```

## 우리동네 달걀 가격
